## Imputing / Fixing the Missing Values for the column labels in the Dataset

```python
## Checking for the datatype of the dataframe
df.dtypes
```

```python
## Imputing / fixing the missing value for Column Label[Age] 
## We are using the function fillna()
## The datatype for Age is float, therefore we will use mean to find the average of Age column
df['Age'].mean()
```

```python
## Assign a variable to the initial code to find the mean
mean_age = df['Age'].mean()      ## the variable name is the mean_age
print(mean_age)
```

```python
## Using the value for the highest_count to fix the missing value
df['Age'] = df['Age'].fillna(mean_age)
```

```python
## Checking if the change has been affected
df.isnull().sum()
```

```python
## Imputing / fixing the missing value for Column Label[KM] 
## We are using the function fillna()
## The datatype for KM is object therefore we will use value_counts() to determine its value KM column
df['KM'].value_counts()
```

```python
## Selecting the hightest value count from KM column
highest_count = df['KM'].value_counts().index[0]  ## Using index to select the highest_count
print(highest_count)
```

```python
## Using the value for the highest_count to fix the missing value
df['KM'] = df['KM'].fillna(highest_count)
```

```python
## Checking if the change has taken place
df.isnull().sum()
```

```python
## Imputing / fixing the missing value for Column Label[FuelType] 
## We are using the function fillna()
## The datatype for FuelType is object therefore we will use value_counts() to determine its value FuelType column
df['FuelType'].value_counts()
```

```python
## Checking if the change has taken place
df.isnull().sum()
```
