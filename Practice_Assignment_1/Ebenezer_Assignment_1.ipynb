```python
# Loading Dataset and Display initial information
import pandas as pd
import numpy as np
df = pd.read_csv('Datasets/Toyota.csv')
df.info()
```

```python
# Display the dataframe
df
```

```python
# Creating a deep copy
df = df.copy(deep=True)
```

```python
# Count NaN values in each column
nan_per_column = df.isna().sum()
print(f'NaN values per column: ')
print(nan_per_column)
```

```python
# Replacing ?? with NaN
df['KM'].replace('??', np.nan)
```

```python
# Replace NaN values in each column with the mean (only for numeric columns)
df = df.fillna(df.mean(numeric_only=True))
```

```python
# Count NaN values in each column
nan_per_column = df.isna().sum()
print(f'NaN values per column: ')
print(nan_per_column)
```

```python
# Display the Dataframe
df
```

```python
# Selecting the values with highest count
df['FuelType'].value_counts()
```

```python
# replacing the values with the highest count
highest_count = df['FuelType'].value_counts().index[0]
```

```python
# replace highest count in the column
df['FuelType'] = df['FuelType'].fillna(highest_count)
```
