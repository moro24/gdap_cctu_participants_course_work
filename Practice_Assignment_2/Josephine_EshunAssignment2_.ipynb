## Data Preprocessing / Cleaning of Dataset
- Normalizing the column names 
- Replacing empty string with nan and fill for missing values 
- deleted the customer id column 
- change the data type on the columns 
- any other processing that is necessary

```python
## Normalizing the column names
## let convert the the column names to lower case
each_items = df.columns
for each_item in df.columns:
    print(each_item.lower())
```

```python
## previewing the columns
df.columns = df.columns.str.lower()
print(df.columns)
```

```python
df.head()
```

```python
## replacing values in totalcharges column 
df['totalcharges'].unique()
```

```python
df["totalcharges"] = df["totalcharges"].replace(0,np.nan)
```

```python
df["totalcharges"]
```

```python
## deleting the customer id column 
df = df.drop("customerid", axis=1)
```

```python
df.head(20)
```

```python
## displaying the first five rows using the transpose
df.head().T   ## the T represent the transpose function and its swaps the rows and columns . 
```

```python
df.dtypes
```

```python
## changing the datatype of 'object' columns to category datatypes.
df["gender"] = df["gender"].astype("category")
df["partner"] = df["partner"].astype("category")
df["dependents"] = df["dependents"].astype("category")
df["phoneservice"] = df["phoneservice"].astype("category")
df["multiplelines"] = df["multiplelines"].astype("category")
df["internetservice"] = df["internetservice"].astype("category")
df["onlinesecurity"] = df["onlinesecurity"].astype("category")
df["onlinebackup"] = df["onlinebackup"].astype("category")
df["deviceprotection"] = df["deviceprotection"].astype("category")
df["techsupport"] = df["techsupport"].astype("category")
df["streamingtv"] = df["streamingtv"].astype("category")
df["streamingmovies"] = df["streamingmovies"].astype("category")
df["contract"] = df["contract"].astype("category")
df["paperlessbilling"] = df["paperlessbilling"].astype("category")
df["paymentmethod"] = df["paymentmethod"].astype("category")
df["totalcharges"] = df["totalcharges"].astype("category")
df["churn"] = df["churn"].astype("category")

df.dtypes
```

```python
## Checking for the counts in the churn column using the value_counts()
df["churn"].value_counts()
```

```python
## replacing the churn values with yes == 1 and no == 0
changing = {'Yes':1, 'No':0}
df["churn"] = df["churn"].replace(changing)
```

```python
df['churn']
```

```python
## preview of churn values
df["churn"].head(20)
```

```python
## checking for churn datatype
df['churn'].dtype
```

```python
## converting churn data types to integer datatype
df['churn'] = df['churn'].astype('int64')
```

## Target Variable Ananlysis

```python
## Using matplotlib to create frequency distribution table on histogram

import matplotlib.pyplot as plt

plt.figure(figsize = (8,4))
plt.hist(df["churn"],color = "red", bins = 10)
plt.title("Frequency Distribution for churn")
plt.xlabel("Churn")
plt.ylabel("Frequency")
plt.show()
```

```python
## Using seaborn to create frequency distribution table on histogram

import seaborn as sns

plt.figure(figsize = (8,4))
sns.histplot(df["churn"],color = "purple", bins = 20)
plt.title("Frequency Distribution for churn")
plt.xlabel("Churn")
plt.ylabel("Frequency")
plt.show()
```

```python
## Using Seaborn to create scatter plot

import seaborn as sns

plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='tenure', y='churn',color='green')

plt.title("Correlation b/n tenure and churn")
plt.xlabel("tenure(months)")
plt.ylabel("churn")

plt.show()
```

```python
## Using Seaborn to create boxplot
plt.figure(figsize=(8,4))
sns.boxplot(x='churn',y='monthlycharges', data=df)

plt.title("Distribution of churn")
plt.xlabel("churn")

plt.show()
```

```python
## Using seaborn for countplot
plt.figure(figsize=(8,4))
sns.countplot(x='seniorcitizen', data=df)

plt.title("Distribution of Senoir Citizen")
plt.xlabel("Senior Citizen")

plt.show()
```
