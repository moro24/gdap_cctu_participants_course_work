# Target Variable Analysis

```python
# Plotting using Seaborn
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
sns.histplot(df['Churn'], bins=10, color='turquoise')
plt.title("Posibilities Of Churn")
plt.xlabel("Churn")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plotting using matplotlib
plt.figure(figsize=(8,4))
plt.hist(df['Churn'], color='black', bins=10)
plt.title("Possibilities Of Churn")
plt.xlabel("Churn")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plotting two variables on a graph
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='Churn', y='gender', color='black')
plt.title("Correlation b/n Churn and gender")
plt.xlabel("Churn")
plt.ylabel("gender")
plt.show()
```

```python
#Plotting a categorical variable
plt.figure(figsize=(8,4))
sns.countplot(x='Churn', data=df)
plt.title("Churn")
plt.xlabel("Churn")
plt.show()
```
