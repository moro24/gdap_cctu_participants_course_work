# Data Preprocessing
- Normalizing the column names
- Replacing empty string with nan and fill for missing values
- deleted the customer id column
- change the data type on the columns
- any other processing that is necessary

```python
# Convert column names to lower case
df.columns = df.columns.str.lower()
```

```python
#preview the columns
df.columns
```

```python
# delete the customer id column 
del df['customerid']
```

```python
df.columns
```

```python
# Display the first five rows with transpose
transposed = df.head().T
print('First five rows (transposed): ')
print(transposed)
```

```python
## lets change the datatype of 'object' columns to category datatypes.
df.dtypes
```

```python
# Convert all object columns to category
# Select columns with dtype 'object'
object_columns = df.select_dtypes(include='object').columns  
df[object_columns] = df[object_columns].astype('category')

print(f'Data Types After Conversion:')
print(df.dtypes)
```

```python
# Convert the target column, where yes == 1 and no = 0
# Convert 'Target' column: 'Yes' -> 1, 'No' -> 0
df['partner'] = df['partner'].replace({'Yes': 1, 'No': 0})
df['dependents'] = df['dependents'].replace({'Yes': 1, 'No': 0})
df['phoneservice'] = df['phoneservice'].replace({'Yes': 1, 'No': 0})
df['onlinesecurity'] = df['onlinesecurity'].replace({'Yes': 1, 'No': 0})
df['onlinebackup'] = df['onlinebackup'].replace({'Yes': 1, 'No': 0})
df['techsupport'] = df['techsupport'].replace({'Yes': 1, 'No': 0})
df['streamingmovies'] = df['streamingmovies'].replace({'Yes': 1, 'No': 0})
df['paperlessbilling'] = df['paperlessbilling'].replace({'Yes': 1, 'No': 0})
df['streamingtv'] = df['streamingtv'].replace({'Yes': 1, 'No': 0})
df['churn'] = df['churn'].replace({'Yes': 1, 'No': 0})
df['deviceprotection'] = df['deviceprotection'].replace({'Yes': 1, 'No': 0})
```
